```python
import pandas as pd
```

```python
vine_df = pd.read_csv('vine_table.csv')
vine_df
```

```python
filtered_df = vine_df[vine_df['total_votes'] >= 20 ]
filtered_df
```

```python
filtered_df['helpful_perc'] = filtered_df['helpful_votes']/filtered_df['total_votes']
filtered_df
```

```python
helpful_reviews = filtered_df[filtered_df['helpful_perc'] >= .5]
helpful_reviews
```

```python
vine_reviews = helpful_reviews[helpful_reviews['vine'] == 'Y']
vine_reviews
```

```python
non_vine_reviews = helpful_reviews[helpful_reviews['vine'] == 'N']
non_vine_reviews
```

```python
total_reviews = [len(vine_reviews.index), len(non_vine_reviews.index)]
star_reviews = [len(vine_reviews[vine_reviews['star_rating'] == 5]), len(non_vine_reviews[non_vine_reviews['star_rating'] == 5])]
star_perc = [len(vine_reviews[vine_reviews['star_rating'] == 5])/len(vine_reviews.index), len(non_vine_reviews[non_vine_reviews['star_rating'] == 5])/len(non_vine_reviews.index)]

print(total_reviews)
print(star_reviews)
print(star_perc)
```

```python
summary_df = pd.DataFrame(
    {'total_reviews': total_reviews,
     '5star_reviews': star_reviews,
     '5star_perc': star_perc,
                             })
print(summary_df.set_axis(['vine_review', 'non_vine_review'], axis='index'))
```
